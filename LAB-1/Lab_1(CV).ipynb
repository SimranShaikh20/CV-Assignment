# Lab 1: Image Basics and Pixel Manipulation

### ðŸ”§ Libraries Used:
- OpenCV (`cv2`)
- NumPy (`numpy`)
- Matplotlib (`matplotlib.pyplot`)




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


color_img = cv2.imread('flower.jpg', cv2.IMREAD_COLOR)


gray_img = cv2.imread('nature.jpg', cv2.IMREAD_GRAYSCALE)
```

# Display both images

```python

rgb_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.title("Color Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Print image shape, datatype, and pixel range

```python
print("Image Shape:", color_img.shape)
print("Data Type:", color_img.dtype)
print("Pixel Range: Min =", color_img.min(), ", Max =", color_img.max())
```

# Change pixel and 50x50 block to red

```python

color_img[50, 50] = [0, 0, 255]


color_img[0:50, 0:50] = [0, 0, 255]

updated_rgb_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)
plt.imshow(updated_rgb_img)
plt.title("Modified Image with Red Block")
plt.axis('off')
plt.show()
```

# Convert image to grayscale manually

```python

manual_gray = np.mean(color_img, axis=2).astype(np.uint8)


plt.imshow(manual_gray, cmap='gray')
plt.title("Manual Grayscale Image")
plt.axis('off')
plt.show()
```

# Split image into B, G, R channels

```python
B, G, R = cv2.split(color_img)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(B, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(G, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(R, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.tight_layout()
plt.show()
```

#  Save the modified image

```python

cv2.imwrite('modified_output.jpg', color_img)
print("Modified image saved as 'modified_output.jpg'")
```

#  Observation

## ðŸ§  Observation:

- OpenCV grayscale uses weighted conversion based on human vision sensitivity (more weight to green).
- Manual grayscale using `np.mean()` equally averages R, G, B â€” result is less realistic.
- The channels show how much each color contributes to the final image.
